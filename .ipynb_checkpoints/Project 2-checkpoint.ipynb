```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
sqlite_file = 'lahman2014.sqlite'
conn = sqlite3.connect(sqlite_file)
```

## Part 1: Wrangling
### Problem 1

Data is pulled from databases using SQL commands. Every columns is pulled from Salaries table. As for Teams, only year, team ID, franchise ID, games played, wins and losses are used.

Since Salaries data begin from 1985, any data from before that is dropped from the teams table.

```python
# Retrieving Salaries and Teams tables from database
salary_query = "SELECT yearID as Year, teamID as Team, salary as Salary FROM Salaries"
winnings_query = "SELECT yearID as Year, teamID as Team, G as 'Games Played', W as Wins FROM Teams"
teams_query = "SELECT teamID as Team FROM Teams"
salaries = pd.read_sql(salary_query, conn)
teams = pd.read_sql(winnings_query, conn)

# list of unique teams (to be used later)
teamID = pd.read_sql(teams_query, conn)
teamID = teamID.drop_duplicates()
teamID = teamID.reset_index(drop=True)

'''
Win percentage computation and dropping data before 1985 (since Salaries 
only has data from 1985 to 2014
'''
teams['Win Percentage'] = (teams['Wins']/teams['Games Played']) * 100
teams = teams[teams['Year'] >= 1985].reset_index(drop=True)

# Computing total payroll for each team in each year
for i,row in teams.iterrows():
    mask = ((salaries['Year'] == row['Year']) & (salaries['Team'] == row['Team']))
    salary_sum = salaries[mask].sum(axis=0)
    teams.at[i,'Total Payroll'] = salary_sum['Salary']
teams.head()
```

## Part 2: Exploratory Data Analysis

### Payroll Distribution

#### Problem 2



```python
plot_data = teams[teams['Year'] >= 1990]

# plotting graph for total payroll over time (values added to max values of year and payroll for padding)
plt.plot(plot_data['Year'],plot_data['Total Payroll'],'b--')
plt.axis([min(plot_data['Year']), max(plot_data['Year'] + 1), 0, max(plot_data['Total Payroll'] + 5000000)])
plt.title("Payroll Across Teams")
plt.xlabel('Year')
plt.ylabel('Total Payroll (in 100M)')

plt.show()
```

#### Question 1
From the graph above, we can see that the payroll range has increased over the years (as in total amount paid). Additionally, as indicated by the lines and the graph below, more teams have increased payrolls by 2014 than earlier years.

#### Problem 3

```python
# get total pay of all teams and count of teams each year
year = 1990
total_payroll = {}
num_teams = {}
std_per_year = {}
while year <= 2014:
    for i,row in plot_data.iterrows():
        if (row['Year'] == year):
            total_payroll.setdefault(row['Year'],0)
            num_teams.setdefault(row['Year'],0)
            std_per_year.setdefault(row['Year'],0)
            total_payroll[year] += row['Total Payroll']
            num_teams[year] += 1
            std_per_year[year] = plot_data[plot_data['Year'] == row['Year']].std()
    year += 1
    
# compute average pay and standard deviation for each year
avg = {}
for key1 in total_payroll: 
    for key2 in num_teams:
        if key1 == key2:
            avg[key1] = [total_payroll[key1]/num_teams[key2],std_per_year[key1]['Total Payroll']]
            
# DataFrame for average payroll every year
avg_std_pay = pd.DataFrame(avg.values(), avg.keys(), columns=['Average', 'Standard Deviation'])

# plotting data from avgerage pay data
plt.plot(avg_std_pay.index + 1, avg_std_pay['Average'], 'b-')
plt.title("Average Payroll Per Year")
plt.ylabel('Avg. Payroll (in 100M)')
plt.xlabel('Year')
plt.show()
```

As shown in the graph above, average payroll had increased over time.

### Correlation between payroll and winning percentage

#### Problem 4

Mean winning percentage and mean total payroll for each team will be calculated as follows: 

A seperate list of unique teams will be used to iterate over and get information from the main DataFrame. A mask will be used to filter out data taht contains the current team, and years from the current year up to (year + 5). Mean will be caculated using pandas.mean() function, then data will be added to a new DataFrame which is appended to period DataFrame. 

Periods will be: 1990-1995, 1996-2000, 2001-2005, 2006, 2010, 2011-2014

```python
# computing average of payrolls and wins for each team for periods of time
year = 1990
period = pd.DataFrame(columns=['Period', 'Team', 'Avg. Win Percentage', 'Avg. Total Payroll'])
while year < 2014:
    for i, row in teamID.iterrows():
        mask = ((teams['Year'] >= year) & (teams['Year'] <= year + 5)) & (teams['Team'] == row['Team'])
        temp = teams[mask]
        temp = temp.reset_index(drop=True)
        if(len(temp) > 0):
            mean_data = temp.mean()
            temp2 = pd.DataFrame([[str(year) + " - " + str(year+5),
                                  temp.at[0,'Team'],
                                  mean_data['Win Percentage'],
                                  mean_data['Total Payroll']]],
                                  columns=['Period', 'Team', 'Avg. Win Percentage','Avg. Total Payroll'])
            period = period.append(temp2, ignore_index=True)
    year += 5
period = period.drop_duplicates()
```

```python
############### Figure (1) ###############

data1 = period[period['Period'] == '1990 - 1995']
plt.figure(1)

for i,row in data1.iterrows():
    if(row['Team'] == 'OAK'):
        plt.scatter(row['Avg. Total Payroll'], row['Avg. Win Percentage'],zorder=3,label=row['Team'],color='red')
        plt.annotate("OAK", xy=(row['Avg. Total Payroll'], row['Avg. Win Percentage'] + 0.5,))
    else:    
        plt.scatter(row['Avg. Total Payroll'], row['Avg. Win Percentage'],label=row['Team'],color='blue')

slope, intercept, r_value, p_value, std_err = stats.linregress(data1['Avg. Total Payroll'], data1['Avg. Win Percentage'])
plt.plot(data1['Avg. Total Payroll'], intercept + slope * data1['Avg. Total Payroll'],color='black',label='Regression Line')

plt.title('1990-1995')
plt.xlabel('Payroll')
plt.ylabel('Winning Percentage')

############### Figure (2) ###############

data2 = period[period['Period'] == '1995 - 2000']
plt.figure(2)

for i, row in data2.iterrows(): 
    if(row['Team'] == 'OAK'):
        plt.scatter(row['Avg. Total Payroll'], row['Avg. Win Percentage'],label=row['Team'], zorder=3,color='red')
        plt.annotate("OAK", xy=(row['Avg. Total Payroll'], row['Avg. Win Percentage'] + 1,))
    else:
        plt.scatter(row['Avg. Total Payroll'], row['Avg. Win Percentage'],label=row['Team'],color='blue')
        
# regression line
slope, intercept, r_value, p_value, std_err = stats.linregress(data2['Avg. Total Payroll'], data2['Avg. Win Percentage'])
plt.plot(data2['Avg. Total Payroll'], intercept + slope * data2['Avg. Total Payroll'],color='black',label='Regression Line')
        
plt.title('1995-2000')
plt.xlabel('Payroll')
plt.ylabel('Winning Percentage')

############### Figure (3) ###############

data3 = period[period['Period'] == '2000 - 2005']
plt.figure(3)

for i, row in data3.iterrows():
    if(row['Team'] == 'OAK'):
        plt.scatter(row['Avg. Total Payroll'], row['Avg. Win Percentage'],label=row['Team'], zorder=3,color='red')
        plt.annotate("OAK", xy=(row['Avg. Total Payroll'], row['Avg. Win Percentage'] + 1,))
    else:
        plt.scatter(row['Avg. Total Payroll'], row['Avg. Win Percentage'],label=row['Team'],color='blue')

# regression line
slope, intercept, r_value, p_value, std_err = stats.linregress(data3['Avg. Total Payroll'], data3['Avg. Win Percentage'])
plt.plot(data3['Avg. Total Payroll'], intercept + slope * data3['Avg. Total Payroll'],color='black',label='Regression Line')
        
plt.title('2000-2005')
plt.xlabel('Payroll')
plt.ylabel('Winning Percentage')

############### Figure (4) ###############

data4 = period[period['Period'] == '2005 - 2010']
plt.figure(4)

for i, row in data4.iterrows():
    if(row['Team'] == 'OAK'):
        plt.scatter(row['Avg. Total Payroll'], row['Avg. Win Percentage'],label=row['Team'], zorder=3,color='red')
        plt.annotate("OAK", xy=(row['Avg. Total Payroll'], row['Avg. Win Percentage'] + 1,))
    else:
        plt.scatter(row['Avg. Total Payroll'], row['Avg. Win Percentage'],label=row['Team'],color='blue')

# regression line
slope, intercept, r_value, p_value, std_err = stats.linregress(data4['Avg. Total Payroll'], data4['Avg. Win Percentage'])
plt.plot(data4['Avg. Total Payroll'], intercept + slope * data4['Avg. Total Payroll'],color='black',label='Regression Line')

plt.title('2005-2010')
plt.xlabel('Payroll')
plt.ylabel('Winning Percentage')

############### Figure (5) ###############

data5 = period[period['Period'] == '2010 - 2015']
plt.figure(5)

for i, row in data5.iterrows():
    if(row['Team'] == 'OAK'):
        plt.scatter(row['Avg. Total Payroll'], row['Avg. Win Percentage'],label=row['Team'], zorder=3,color='red')
        plt.annotate("OAK", xy=(row['Avg. Total Payroll'], row['Avg. Win Percentage'] + 1,))
    else:
        plt.scatter(row['Avg. Total Payroll'], row['Avg. Win Percentage'],label=row['Team'],color='blue')

# regression line
slope, intercept, r_value, p_value, std_err = stats.linregress(data5['Avg. Total Payroll'], data5['Avg. Win Percentage'])
plt.plot(data5['Avg. Total Payroll'], intercept + slope * data5['Avg. Total Payroll'],color='black',label='Regression Line')
        
plt.title('2010-2015')
plt.xlabel('Payroll')
plt.ylabel('Winning Percentage')

###########################################

plt.show()
```

### Standardizing across years
#### Problem 5

```python
# adding mean payroll and standard deviation to original DataFrame
teams['Avg. Payroll Per Year'] = 0
teams['Year Payroll SD.'] = 0
for i, row in teams.iterrows():
    if(row['Year'] in avg_std_pay.index):
        teams.at[i,'Avg. Payroll Per Year'] = avg_std_pay.at[row['Year'], 'Average']
        teams.at[i,'Year Payroll SD.'] = avg_std_pay.at[row['Year'], 'Standard Deviation']

# computing standaradized payroll for each team
teams['Standardized Payroll'] = (teams['Total Payroll'] - teams['Avg. Payroll Per Year'])/(teams['Year Payroll SD.'])
plot_data = teams[teams['Year'] >= 1990]
plot_data.head()
```
